## 1. What is BigML

BigML is providing a selection of robust Machine Learning algorithms to solve real world problems by applying a single, standardized framework across your company or organization. BigML facilitates unlimited predictive applications across industries and using BigML many of your dependencies and complexity has been decreased using BigML APIs.

To Keep it short, it is a platform where you can build your own machine learning models without knowing much detail of implementation. They are providing APIs and using those APIs you can build entire ML model and also they have platforms available for the deployment as well. We will see step by step process to build sample model using BigML Platform.

Before going to implementation part, Let's ook at some of the features:-

- **Comprehensive Machine Learning Platform** to solve real world problems by applying a single, standardized framework across your company.


- **Immediate Access** - you will be up and running with an easy-to-use web interface and REST API in a matter of seconds. They have free as well as paid plans for handling small as well as large/complex data sets.


- **Interpretable & Exportable Models** - This will be the great feature because All predictive models on BigML come with an interactive visualization and explainability features that make them interpretable. They can be exported and used to serve offline predictions on any edge computing device or be instantaneously deployed as part of real-time production applications.


- **Collaboration** - You can share your Machine Learning resources using team and project management capabilities of BigML.


- **Programmable & Repeatable** - BigML brings every new feature first to the REST API. Bindings & libraries are available for all popular languages, including Python, Node.js, Ruby, Java, Swift, and more. This will be the great fetaure for developer as to use this REST APIs.


- **Automation** - Rapidly bring your predictive modeling tasks to production through effective automation as BigML providing such kind of automation for executing complex workflows.


- **Flexible Deployments** - BigML give options for multi-tenant and single-tenant versions on the cloud or on-premises. BigML can be ported to any cloud provider or to a Virtual Private Cloud, with fully-managed and self-managed versions.


All of these features are deeply mentioned at their site. For more detail about it you can refer here:- 
https://bigml.com/features

## 2. Creating Model using BigML

In This section, we will see the implementation of creating model in BigML. This will help to create source, dataset, model and prediction in BigML.

You need some of the prerequisite to start with BigMl.

- Create an account in BigML(https://bigml.com/accounts/register/)


- A terminal with curl or any other command-line tool that implements standard HTTPS methods. For this notebook, we are using curl and the curl command has been given for windows as well as for pther OS like unix, linux,MAC (Same command for all OS ,slight difference in windows command). Try those commands depend on your operating system.


- Need some sample data. here we are using famous flower iris dataset. You can download from here:- https://static.bigml.com/csv/iris.csv  else you can find in this folder as well.


Below are the steps to be follow to create and predict the model in BigML:-


<a id="data"></a>
### 2.1 Getting iris data

```curl -o iris.csv https://static.bigml.com/csv/iris.csv```

This will download the iris data from the url and save in the file iris.csv

<a id="authentication"></a>
### 2.2 Authentication

When you create BigML account,an API key has been generated for your account. While accessing API,some kind of authentication is required. Below is the code to set up an enviornment variable for your authentication/APi-Key.

```export BIGML_USERNAME=Your_User_name
export BIGML_API_KEY=Your_API_Key
export BIGML_AUTH="username=$BIGML_USERNAME;api_key=$BIGML_API_KEY"```


If you are using windows OS, then below is the code:-

```set BIGML_USERNAME=Your_User_name
set BIGML_API_KEY=Your_API_Key
set BIGML_AUTH=username^=%BIGML_USERNAME%;api_key^=%BIGML_API_KEY%```

For more about authentication, you will find here:- https://bigml.com/api/authentication

### 2.3 Creating a Model

For creating new Model, POST the dataset/id from the previous step to the model base URL. By default BigML.io will include all fields as predictors and will treat the last non-text field as the objective.

```curl "https://bigml.io/andromeda/model?$BIGML_AUTH" \
    -X POST \
    -H 'content-type: application/json' \
    -d '{"dataset": "dataset/4f52da4303ce896fe3000000"}'```

For windows user:-

```curl "https://bigml.io/andromeda/model?%BIGML_AUTH%" ^
    -X POST ^
    -H "content-type: application/json" ^
    -d "{\"dataset\": \"dataset/4f52da4303ce896fe3000000\"}"```

## 3. BigMLer

BigMLer is the command line tool for Machine Learning. BigMLer is a free and open source tool that will help you perform sophisticated Machine Learning workflows by typing just one line in your command prompt.

Here is the documentation you can refer for more detail:- https://bigmler.readthedocs.io/en/latest/

The same thing which we have done till now we can do the same thing using this tool in one line sentense as well. Refer this doc for that implementation.

There is a need for installation of BigMLer given in the doc, install using pip or any other method. Basic command is:-

`pip install bigmler`

Before going to retraining of model using BigMler, there is one thing to note here:-

In BigML, once a model (or any other resource) is built, it is immutable. None of its essential properties can be changed (only non-essential attributes, like their name or description, can be updated). The model keeps the information about what resources were used to build it or which configuration choices were made to get it. That ensures that all the steps that led to your actual model can be deduced i.e, the model-building workflow. By simply applying this workflow to new data you can get a retrained model.

Hence, retraining of model using new data can be done with new data in a single line command:-

<a id="BigMLer_retraining"></a>
## 4. Retraining BigMLer command

```bigmler retrain --id model/5a3ae0f14006833a070003a4 --add data/iris.csv \
                --output-dir retrain_directory```
                
                
so that the data in your local data/iris.csv file is uploaded to the platform and all the steps that led to your existing model are reproduced to create a new merged dataset that will be used to retrain your model. The command output will contain the URL that you need to call to ensure you always use the latest version of your model. here is the output:-

https://bigml.io/andromeda/ensemble?username=my_user;api_key=my_api_key;limit=1;full=yes;tags=my_ensemble



In a different scenario, you might want to retrain your model from a list of datasets, for instance training an anomaly detector using the data of the last 6 months. This means that you don’t want your data to be merged. Rather you would like to use a window over the list of available datasets.

```bigmler retrain --ensemble-tag my_ensemble --add data/iris.csv \
                --window-size 6 --output-dir retrain_directory```
                
In this case, adding the --window-size option to your command will cause the dataset created by uploading your new data to be added to the list of datasets as a separate resource. Then model will be rebuilt using the number of datasets set as --window-size.
